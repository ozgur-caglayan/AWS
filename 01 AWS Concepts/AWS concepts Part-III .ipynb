# AWS Concepts

## AWS Global Infrastructure

*  Global infrastructure meaning we're going to talk about how physically AWS is set up and deployed around the globe as per usual.

* This map shows the locations of all of the AWS regions. So what are AWS regions?
    * So each region is a geographical area.
    * That is a collection of AWS availability zones and data centers.
    

* So just give you an idea of why there is all of these regions spread out across the world is specifically this.
    * If somebody is in Tokyo and they're using AWS and they want to provision and use an EC2 instance.
    * They want to be able to launch that EC2 instance on a physical hardware that is in a data center in Tokyo.
    * Because they want that physical hardware to be close or there is a lot less latency while transmitting data.
    * Whereas if they had to provision and use an EC2 instance say that was in North Virginia any data that they were to send back and forth would have to travel halfway around the world and back.
    * That's taking a lot more time and having a lot more transmission latency.


* So AWS to be fair to all of their customers around the world and to provide the absolute best performance has put regions and data centers and physical hardware all around the globe.
* There's currently 11 active consumer based regions with one government base regions being the AWS of cloud.
* There is also four new regions which are currently under development at the time of this recording which is in September of 2016.


* So now what we're going to do is we're going to zoom in on one of these regions to see what's going.
* So let's hypothetically say we're zooming in the north Virginia region.

* So once we zoom into an AWS region, we see that is made up of **availability zones**.
* Each region may have a different number of availabilities zones. But for this example let's say that this particular region has three availabilities zones.


* Now what is an availability zone?
    * Availability zone is actually a geographical physical location that holds in AWS data center.
    * So now we've zoomed into an availability zone and there was where we see the physical hardware the physical data center where all of the AWS resources and any data or information that you put on AWS is located.

* So maybe zoom back out again and we'll go through all the regions across the globe that AWS has set up. Each region has a set of availability zones.
* Now each one of these availabilities zones contains a data center and each one of these availabilities zones is geographically separated from the other.
    * So if this is the North Virginia region there is an availability zone with a data center somewhere north Virginia.
    * And then some miles away whether it's 20 miles 50 miles 100 miles somewhere else in that region of the country, there is another availability zone.
    * In most cases there's also a third and sometimes a fourth availability zone.


* Now why did they have all these various availabilities zones with in a particular AWOS region?
    * And that again goes back to high availability and fault tolerance.
    * In that let's say that there was a natural disaster somewhere in North Virginia.
    * There was a major earthquake or power outage or something that actually knocked out one of these data centers within a region.
    * There is redundancy across these availabilities zones so if availability zone one were to fail or go down, most if not all of your files have already been backed up in "availability Zone 2",  "an availability Zone 3" within this particular region.
    * So as long as it doesn't knock out the entire region then you still should be able to access AWS resources and any files or information have uploaded and use AWS.
    
    
* So again when we zoom into one of these availabilities zones we see that the availability zone is where the physical hardware that makes up the data center is located.
* It is inside of these servers, inside of these racks, inside of these computers here, where this is located right where your VPC is located where when you actually launch an EC2 instance or you place a file inside of an Amazon S3 bucket will back out now.
* Now this is all taking place with in a data center which is located within one particular availabilities zone which is located within a region that AWS has set up.


* We have our regions,
* Within our regions we have availability zones and,
* There are multiple availabilities zones for redundancy.
* And then inside each availability zone our data centers the physical hardware and that it is those physical pieces of hardware that you create and provision your AWS resources or utilize something like an S3 bucket.


* So when you create an AWS account, and you go to create a bucket, or to launch an Amazon EC2 instance; you're always choosing what region and what availability zone you want those things take place.
* So I hope that explanation gives you a good overarching understanding of the general concepts of how AWS is set up globally and how their infrastructure is viewed from a very high level and then paring it down to a very local level.
