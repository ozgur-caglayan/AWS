## AWS Global Infrastructure

*  Global infrastructure meaning we're going to talk about how physically AWS is set up and deployed around the globe as per usual.

* This map shows the locations of all of the AWS regions. So what are AWS regions?
    * So each region is a geographical area.
    * That is a collection of AWS availability zones and data centers.
    

* So just give you an idea of why there is all of these regions spread out across the world is specifically this.
    * If somebody is in Tokyo and they're using AWS and they want to provision and use an EC2 instance.
    * They want to be able to launch that EC2 instance on a physical hardware that is in a data center in Tokyo.
    * Because they want that physical hardware to be close or there is a lot less latency while transmitting data.
    * Whereas if they had to provision and use an EC2 instance say that was in North Virginia any data that they were to send back and forth would have to travel halfway around the world and back.
    * That's taking a lot more time and having a lot more transmission latency.


* So AWS to be fair to all of their customers around the world and to provide the absolute best performance has put regions and data centers and physical hardware all around the globe.
* There's currently 11 active consumer based regions with one government base regions being the AWS of cloud.
* There is also four new regions which are currently under development at the time of this recording which is in September of 2016.
